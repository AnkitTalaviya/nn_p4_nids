# 5-4-1 Offline Training Notebook

Minimal notebook for the final offline training/export path of the **5-4-1** model.

**Feature set:** `proto, sttl, dttl, swin, dwin`


```python
from pathlib import Path
import json
from train_5_4_1_offline import FEATURES, run_pipeline

code_dir = Path.cwd().resolve()
if not (code_dir / 'train_5_4_1_offline.py').exists():
    code_dir = (Path.cwd().resolve() / '5-4-1/offline/python/code').resolve()

repo_root = code_dir.parents[3]
train_csv = repo_root / 'data/UNSW_NB15_training-set.csv'
test_csv = repo_root / 'data/UNSW_NB15_testing-set.csv'
output_dir = code_dir.parent / 'output'

result = run_pipeline(
    train_csv=train_csv,
    test_csv=test_csv,
    output_dir=output_dir,
    random_state=42,
    max_iter=80,
)

metrics = json.loads(Path(result['metrics_path']).read_text(encoding='utf-8'))

print('Architecture: 5-4-1')
print('Features:', FEATURES)
print('\nFloat metrics:')
print(metrics['float'])
print('\nQuantized metrics:')
print(metrics['quantized'])
print('\nThreshold:', metrics['threshold'])
print('Logit correlation:', metrics.get('logit_correlation'))
print('\nArtifacts:')
print(' -', result['model_path'])
print(' -', result['metrics_path'])
print(' -', result['predictions_path'])

result
```
